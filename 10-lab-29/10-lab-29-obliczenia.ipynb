```python
# fmt: off
import numpy as np
import matplotlib.pyplot as plt

# Temperatura w stopniach Celsjusza
T = np.array([
    11.0, 10.8, 10.6, 10.4, 10.2, 10.0, 9.8, 9.6, 9.4, 9.2, 9.0, 8.8, 8.6, 8.4, 8.2,
    8.0, 7.8, 7.6, 7.4, 7.2, 7.0, 6.8, 6.6, 6.4, 6.2, 6.0, 5.8, 5.6, 5.4, 5.2,
    5.0, 4.8, 4.6, 4.4, 4.2, 4.0, 3.8, 3.6, 3.4, 3.2, 3.0, 2.8, 2.6, 2.4, 2.2,
    2.0, 1.8, 1.6, 1.4, 1.2, 1.0, 0.8, 0.6, 0.4, 0.2
]) 

# Wysoko supa wody w mm - seria 1
h1 = np.array([
    80, 77, 75, 74, 72, 69, 68, 66, 64, 63, 61, 61, 60, 56, 55,
    53, 52, 51, 50, 49, 47, 46, 45, 44, 43, 42, 42, 42, 40, 40,
    39, 38, 38, 38, 37, 37, 37, 37, 37, 37, 37, 37, 37, 38, 38,
    38, 39, 39, 40, 40, 41, 41, 42, 43, 43
]) * 1e-3

# Wysoko supa wody w mm - seria 2
h2 = np.array([
    71, 70, 68, 66, 64, 62, 60, 58, 56, 55, 54, 51, 51, 49, 48,
    46, 44, 44, 43, 41, 40, 39, 37, 37, 36, 40, 36, 34, 32, 32,
    30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 32,
    33, 34, 34, 35, 36, 37, 38, 39, 40, 41
]) * 1e-3
# fmt: on

d = 0.0017

V1 = np.pi * d**2 / 4 * h1
V2 = np.pi * d**2 / 4 * h2

Delta_V1 = V1 - V1[0]
Delta_V2 = V2 - V2[0]

print("V1:")
print(V1)
print("V2:")
print(V2)
print("Delta V1:")
print(Delta_V1)
print("Delta V2:")
print(Delta_V2)

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot Series 1
(line1,) = ax1.plot(T, h1, "b-", label="Wysoko", linewidth=2)
ax1.set_xlabel("Temperatura [$^\\circ$C]", fontsize=12)
ax1.set_ylabel("Wysoko [m]", fontsize=12, color="black")
ax1.tick_params(axis="y", labelcolor="black")
ax1.grid(True, linestyle="--", alpha=0.7)
ax1.set_title("Seria 1", fontsize=14)

ax1_twin = ax1.twinx()
(line3,) = ax1_twin.plot(T, Delta_V1, "b--", label="$\Delta V$", linewidth=1.5)
ax1_twin.set_ylabel("$\Delta V$ [m$^3$]", fontsize=12, color="black")
ax1_twin.tick_params(axis="y", labelcolor="black")

# Plot Series 2
(line2,) = ax2.plot(T, h2, "r-", label="Wysoko", linewidth=2)
ax2.set_xlabel("Temperatura [$^\\circ$C]", fontsize=12)
ax2.set_ylabel("Wysoko [m]", fontsize=12, color="black")
ax2.tick_params(axis="y", labelcolor="black")
ax2.grid(True, linestyle="--", alpha=0.7)
ax2.set_title("Seria 2", fontsize=14)

ax2_twin = ax2.twinx()
(line4,) = ax2_twin.plot(T, Delta_V2, "r--", label="$\Delta V$", linewidth=1.5)
ax2_twin.set_ylabel("$\Delta V$ [m$^3$]", fontsize=12, color="black")
ax2_twin.tick_params(axis="y", labelcolor="black")

# Invert x-axis for both plots
ax1.invert_xaxis()
ax2.invert_xaxis()

# Add legends
lines1 = [line1, line3]
labels1 = [line.get_label() for line in lines1]
ax1.legend(lines1, labels1, loc="best", fontsize=10)

lines2 = [line2, line4]
labels2 = [line.get_label() for line in lines2]
ax2.legend(lines2, labels2, loc="best", fontsize=10)

plt.suptitle(
    "Wysoko supa wody i zmiana objtoci w zalenoci od temperatury", fontsize=14
)
plt.tight_layout()
plt.show()
```
