```python
import matplotlib.pyplot as plt
import numpy as np

# --- KROK 1: Dane pomiarowe z Twojej tabeli ---
# Dane zostały posortowane rosnąco według odległości,
# aby wykres był narysowany w prawidłowej kolejności.

# Położenie mikrofonu 'x' w centymetrach [cm]
x_cm = [80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90]

# Zmierzone napięcie 'U' w woltach [V]
U_V = [0.084, 0.074, 0.058, 0.045, 0.029, 0.025, 0.016, 0.029, 0.046, 0.061, 0.071]


# --- KROK 2: Generowanie wykresu ---
# Ta część kodu jest uniwersalna i nie wymaga zmian.

# Stworzenie okna wykresu o odpowiednich wymiarach
plt.figure(figsize=(12, 7))

# Narysowanie punktów pomiarowych (jako kółka 'o')
# i połączenie ich ciągłą linią ('-')
plt.plot(x_cm, U_V, 'o-', color='crimson', label='Dane pomiarowe')

# Zaznaczenie na wykresie punktu o minimalnym napięciu (węzła)
min_voltage = min(U_V)
min_pos_index = U_V.index(min_voltage)
min_x = x_cm[min_pos_index]
plt.plot(min_x, min_voltage, 's', color='blue', markersize=10, label=f'Węzeł fali (x={min_x} cm)')


# Dodanie tytułu i etykiet osi z jednostkami
plt.title('Wykres zależności napięcia U od położenia mikrofonu x (dla f = 800 Hz)', fontsize=16)
plt.xlabel('Położenie mikrofonu x [cm]', fontsize=12)
plt.ylabel('Napięcie U [V]', fontsize=12)

# Włączenie siatki dla łatwiejszego odczytu
# Ustawienie minimalnej wartości osi Y na 0 dla lepszej czytelności
plt.grid(True, linestyle='--', alpha=0.7)

plt.ylim(bottom=0)

# Dodanie legendy
plt.legend(fontsize=12)

# Wyświetlenie gotowego wykresu
plt.show()
```
