# 17. lab 73 - obliczenia


```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
delta_x_max = 0.5
delta_u_max = 0.001

u_x = delta_x_max / np.sqrt(3)
u_u = delta_u_max / np.sqrt(3)

print(f"Standard Uncertainty u(x): {u_x:.4f} cm")
print(f"Standard Uncertainty u(U): {u_u:.4f} V")
```

```python
x_cm = [80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90]
U_V = [0.084, 0.074, 0.058, 0.045, 0.029, 0.025, 0.016, 0.029, 0.046, 0.061, 0.071]

plt.figure(figsize=(12, 7))

plt.errorbar(
    x_cm,
    U_V,
    xerr=u_x,
    yerr=u_u,
    fmt="o",
    color="crimson",
    ecolor="black",
    elinewidth=1.5,
    capsize=2,
    markersize=2,
    label="Napicie [V] u(x), u(U)$"
)

min_voltage = min(U_V)
min_pos_index = U_V.index(min_voltage)
min_x = x_cm[min_pos_index]

plt.plot(
    min_x,
    min_voltage,
    "s",
    color="blue",
    markersize=3,
    label=f"Wze fali (x={min_x} cm)",
)

plt.title(
    "Zaleno napicia U od pooenia mikrofonu x (f = 800 Hz).",
    fontsize=16,
)
plt.xlabel("Pooenie mikrofonu x [cm]", fontsize=12)
plt.ylabel("Napicie U [V]", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.7)

plt.ylim(bottom=0)
plt.legend(fontsize=12)

output_path = "17-lab-73/report/img/u_x.png"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

plt.savefig(output_path, dpi=600, bbox_inches='tight')
print(f"Figure saved to: {output_path}")

plt.show()
```
