```python
import numpy as np


def plot_temperature_analysis(df, series_name, interpolation_time=300, n=10):
    """
    Analyze and plot temperature data with trend lines.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing time and temperature data
    series_name : str
        Name of the temperature series to analyze (e.g., 'T1', 'T2', 'T3')
    interpolation_time : int, optional
        Time point for temperature interpolation (default: 300)
    n : int, optional
        Number of points to use for trend line fitting (default: 10)
    """
    # Get time and temperature slices
    time_slice_start = df["t"].head(n)
    temp_slice_start = df[series_name].head(n)
    time_slice_end = df["t"].tail(n)
    temp_slice_end = df[series_name].tail(n)

    # Fit lines to the data
    slope_start, intercept_start = np.polyfit(time_slice_start, temp_slice_start, 1)
    slope_end, intercept_end = np.polyfit(time_slice_end, temp_slice_end, 1)

    # Calculate interpolated temperatures at interpolation_time
    t_p = (
        slope_start * interpolation_time + intercept_start
    )  # Temperature at interpolation_time from start line
    t_k = (
        slope_end * interpolation_time + intercept_end
    )  # Temperature at interpolation_time from end line

    print(f"Temperatura pocztkowa (t_p): {t_p:.2f}C")
    print(f"Temperatura kocowa (t_k): {t_k:.2f}C")

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(df["t"], df[series_name], marker="o", color="tab:red", label=series_name)
    plt.plot(
        df["t"],
        slope_start * df["t"] + intercept_start,
        "--",
        color="blue",
        label="Linia trendu (pocztek)",
    )
    plt.plot(
        df["t"],
        slope_end * df["t"] + intercept_end,
        "--",
        color="green",
        label="Linia trendu (koniec)",
    )
    plt.axvline(
        x=interpolation_time,
        color="gray",
        linestyle="--",
        label=f"Wrzucenie ciaa ({interpolation_time//60:02d}:{(interpolation_time%60):02d})",
    )
    plt.title(f"Temperatura {series_name} w czasie")
    plt.xlabel("Czas [s]")
    plt.ylabel("Temperatura [C]")
    plt.grid(True)
    plt.legend()
    plt.show()

    return t_p, t_k
```

```python
plot_temperature_analysis(df, "T1")

plot_temperature_analysis(df, "T2", interpolation_time=330)

plot_temperature_analysis(df, "T3")
```

## Ciepa waciwe

$$
    c_p = \frac{[m_w \cdot c_w + m_n \cdot c_n] \cdot (T_p - T_k)}{m_c \cdot (T_k - T_c)}
$$


```python
Tp1 = 25.30  # Temperatura pocztkowa dla ciaa miedzianego
Tk1 = 30.75  # Temperatura kocowa dla ciaa miedzianego

Tp2 = 24.90  # Temperatura pocztkowa dla ciaa mosinego
Tk2 = 30.40  # Temperatura kocowa dla ciaa mosinego

Tp3 = 25.10  # Temperatura pocztkowa dla ciaa aluminiowego
Tk3 = 27.78  # Temperatura kocowa dla ciaa aluminiowego

Tc1 = 100.2
Tc2 = 100.4
Tc3 = 102.4

m_w1 = 0.0668
m_w2 = 0.0612
m_w3 = 0.0756

m_n = 0.1261
c_w = 4186
c_n = 377
m_1 = 0.073
m_2 = 0.0708
m_3 = 0.0152


def cieplo_wlasciwe(Tp, Tk, Tc, m_w, m_n, c_w, c_n, m_c):
    return (m_w * c_w + m_n * c_n) * (Tp - Tk) / (m_c * (Tk - Tc))


c1 = cieplo_wlasciwe(Tp1, Tk1, Tc1, m_w1, m_n, c_w, c_n, m_1)
c2 = cieplo_wlasciwe(Tp2, Tk2, Tc2, m_w2, m_n, c_w, c_n, m_2)
c3 = cieplo_wlasciwe(Tp3, Tk3, Tc3, m_w3, m_n, c_w, c_n, m_3)

print(c1)
print(c2)
print(c3)

c1_target = 385
c2_target = 375
c3_target = 900

print(c1_target - c1)
print(c2_target - c2)
print(c3_target - c3)


def relative_difference(measured, target):
    """
    Oblicza wzgldn rnic midzy wartoci zmierzon a docelow.

    Parametry:
    -----------
    measured : float
        Warto zmierzona
    target : float
        Warto docelowa

    Zwraca:
    --------
    float
        Wzgldna rnica w procentach
    """
    return abs((measured - target) / target) * 100


print(f"Wzgldna rnica dla miedzi: {relative_difference(c1, c1_target):.2f}%")
print(f"Wzgldna rnica dla mosidzu: {relative_difference(c2, c2_target):.2f}%")
print(f"Wzgldna rnica dla aluminium: {relative_difference(c3, c3_target):.2f}%")
```

```python
def cieplo_wlasciwe_z_niepewnoscia(
    Tp, Tk, Tc, m_w, m_n, c_w, c_n, m_i, niepewnosci=None
):
    """
    Oblicza ciepo waciwe ciaa staego oraz jego niepewno.
    Wartoci c_w i c_n s traktowane jako dokadne (odczytane z tabel).

    Parametry:
    -----------
    Tp : float
        Temperatura pocztkowa wody (C)
    Tk : float
        Temperatura kocowa ukadu (C)
    Tc : float
        Temperatura pocztkowa badanego ciaa (C)
    m_w : float
        Masa wody (kg)
    m_n : float
        Masa naczynia kalorymetrycznego (kg)
    c_w : float
        Ciepo waciwe wody (J/kgK) - z tablic
    c_n : float
        Ciepo waciwe naczynia (J/kgK) - z tablic
    m_i : float
        Masa badanego ciaa (kg)
    niepewnosci : dict, optional
        Sownik zawierajcy niepewnoci dla kadego parametru:
        {'Tp': float, 'Tk': float, 'Tc': float, 'm_w': float,
         'm_n': float, 'm_i': float}

    Zwraca:
    --------
    tuple : (float, float lub None)
        Ciepo waciwe (J/kgK) i jego niepewno (jeli podano niepewnoci)
    """
    # Obliczanie zmian temperatury
    delta_T = Tp - Tk  # Zmiana temperatury wody + naczynia
    delta_T_i = Tk - Tc  # Zmiana temperatury badanego ciaa

    # Obliczanie iloczynu pojemnoci cieplnej
    Q = m_w * c_w + m_n * c_n

    # Obliczanie ciepa waciwego
    c_i = (Q * delta_T) / (m_i * delta_T_i)

    # Obliczanie niepewnoci, jeli podano
    niepewnosc = None
    if niepewnosci:
        # Pochodne czstkowe zgodnie z podanymi wzorami
        d_c_i_d_m_w = (c_w * delta_T) / (m_i * delta_T_i)
        d_c_i_d_m_n = (c_n * delta_T) / (m_i * delta_T_i)
        d_c_i_d_delta_T = Q / (m_i * delta_T_i)
        d_c_i_d_m_i = -c_i / m_i
        d_c_i_d_delta_T_i = -c_i / delta_T_i

        # print(f"c_i/m_w = (c_w * T) / (m_i * T_i) = {d_c_i_d_m_w:.6f}")
        # print(f"c_i/m_n = (c_n * T) / (m_i * T_i) = {d_c_i_d_m_n:.6f}")
        # print(f"c_i/T = Q / (m_i * T_i) = {d_c_i_d_delta_T:.6f}")
        # print(f"c_i/m_i = -c_i / m_i = {d_c_i_d_m_i:.6f}")
        # print(f"c_i/T_i = -c_i / T_i = {d_c_i_d_delta_T_i:.6f}")

        # Obliczanie skadnikw niepewnoci
        u_m_w = abs(d_c_i_d_m_w) * niepewnosci.get("m_w", 0)
        u_m_n = abs(d_c_i_d_m_n) * niepewnosci.get("m_n", 0)

        # Dla niepewnoci temperatur
        u_Tp = abs(d_c_i_d_delta_T) * niepewnosci.get("Tp", 0)
        u_Tk_delta_T = abs(d_c_i_d_delta_T) * niepewnosci.get("Tk", 0)
        u_Tk_delta_T_i = abs(d_c_i_d_delta_T_i) * niepewnosci.get("Tk", 0)
        u_Tc = abs(d_c_i_d_delta_T_i) * niepewnosci.get("Tc", 0)

        u_m_i = abs(d_c_i_d_m_i) * niepewnosci.get("m_i", 0)

        # Cakowita niepewno (metoda bdu maksymalnego)
        niepewnosc = u_m_w + u_m_n + u_Tp + u_Tk_delta_T + u_Tk_delta_T_i + u_Tc + u_m_i

    return c_i, niepewnosc
```

```python
niepewnosci = {
    "Tp": 0.01,  # K (niepewno termometru)
    "Tk": 0.01,  # K (niepewno termometru)
    "Tc": 0.01,  # K (niepewno termometru)
    "m_w": 0.0001,  # kg (niepewno wagi)
    "m_n": 0.0001,  # kg (niepewno wagi)
    "m_i": 0.0001,  # kg (niepewno wagi)
}

# Obliczenie ciepa waciwego i niepewnoci dla wszystkich cia
c1, delta_c1 = cieplo_wlasciwe_z_niepewnoscia(
    Tp1, Tk1, Tc1, m_w1, m_n, c_w, c_n, m_1, niepewnosci
)

c2, delta_c2 = cieplo_wlasciwe_z_niepewnoscia(
    Tp2, Tk2, Tc2, m_w2, m_n, c_w, c_n, m_2, niepewnosci
)

c3, delta_c3 = cieplo_wlasciwe_z_niepewnoscia(
    Tp3, Tk3, Tc3, m_w3, m_n, c_w, c_n, m_3, niepewnosci
)

print(f"Ciepo waciwe ciaa 1 (mied): {c1:.2f}  {delta_c1:.2f} J/(kgK)")
print(f"Ciepo waciwe ciaa 2 (mosidz): {c2:.2f}  {delta_c2:.2f} J/(kgK)")
print(f"Ciepo waciwe ciaa 3 (aluminium): {c3:.2f}  {delta_c3:.2f} J/(kgK)")
```
