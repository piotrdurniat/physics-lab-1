# 00. lab 00 - obliczenia


```python
import numpy as np

# dane z tabeli (jedyny prawdziwy zestaw)
data = {
    'R1': np.array([97, 52, 27.6]),
    'R2': np.array([202, 90, 49.3]),
    'R3': np.array([832, 452, 232]),
    'R4': np.array([2320, 1340, 653]),
    'R5': np.array([6100, 3299, 1720]),
    'Rr': np.array([64, 32, 16.4]),
    'Rs': np.array([1060, 560, 309])
}
# stosunki l1/l2 dla L=1/3, 1/2, 2/3
ratios = np.array([0.5, 1.0, 2.0])

res = {} # wyniki srednie
unc = {} # niepewnosci

print(f"{'R':<3} | {'L=1/3':<7} {'L=1/2':<7} {'L=2/3':<7} | {'Avg':<7} {'Unc':<5}")
print("-" * 50)

for k, v in data.items():
    Rx = v * ratios
    res[k] = np.mean(Rx)
    unc[k] = (np.max(Rx) - np.min(Rx)) / 2
    print(f"{k:<3} | {Rx[0]:<7.1f} {Rx[1]:<7.1f} {Rx[2]:<7.1f} | {res[k]:<7.1f} {unc[k]:<5.1f}")

# teoria
Rs_theo = res['R1'] + res['R2'] + res['R3']
Rr_theo = 1 / (1/res['R1'] + 1/res['R2'] + 1/res['R3'])

print(f"\nRs teor: {Rs_theo:.1f}, pomiar: {res['Rs']:.1f}, diff: {abs(res['Rs']-Rs_theo):.1f}")
print(f"Rr teor: {Rr_theo:.1f}, pomiar: {res['Rr']:.1f}, diff: {abs(res['Rr']-Rr_theo):.1f}")
```
