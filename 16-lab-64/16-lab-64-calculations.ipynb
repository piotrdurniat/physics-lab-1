```python
import numpy as np


# Konwersja stopni i minut na stopnie dziesiętne
def deg_min_to_dec(degrees, minutes):
    return degrees + minutes / 60.0


# Numer prążka
n = np.array([0, 1, 2, 3, 4])

# Kąty dla lewej krawędzi (L) w stopniach dziesiętnych
# np.nan oznacza brak pomiaru (--)
phi_1L = np.array(
    [
        deg_min_to_dec(353, 40),
        deg_min_to_dec(351, 5),
        deg_min_to_dec(348, 20),
        deg_min_to_dec(345, 35),
        np.nan,
    ]
)

phi_2L = np.array(
    [
        deg_min_to_dec(173, 50),
        deg_min_to_dec(171, 5),
        deg_min_to_dec(168, 20),
        deg_min_to_dec(165, 35),
        # np.nan
    ]
)

# Kąty dla prawej krawędzi (P) w stopniach dziesiętnych
phi_1P = np.array(
    [
        np.nan,
        deg_min_to_dec(359, 10),
        deg_min_to_dec(1, 50),
        deg_min_to_dec(4, 39),
        # deg_min_to_dec(7, 35)
    ]
)

phi_2P = np.array(
    [
        np.nan,
        deg_min_to_dec(179, 6),
        deg_min_to_dec(181, 35),
        deg_min_to_dec(184, 39),
        # deg_min_to_dec(187, 30)
    ]
)
# print(phi_1L[1:])

phi_all = zip(phi_1L, phi_2L, phi_1P, phi_2P)

# print(phi_all)
phi_bar = np.array(
    [
        (phi_1L_i - phi_1P_i + phi_2L_i - phi_2P_i) / 2
        for (phi_1L_i, phi_2L_i, phi_1P_i, phi_2P_i) in phi_all
    ]
)

print(phi_bar)

# Wyświetlenie tablic (do Twojej wiadomości, możesz to usunąć)
# print("Numer prążka (n):")
# print(n)
# print("\nKąt phi_1L [stopnie]:")
# print(phi_1L)
# print("\nKąt phi_2L [stopnie]:")
# print(phi_2L)
# print("\nKąt phi_1P [stopnie]:")
# print(phi_1P)
# print("\nKąt phi_2P [stopnie]:")
# print(phi_2P)
```
