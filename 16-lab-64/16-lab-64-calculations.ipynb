`````python
import numpy as np

def angle_diff(phi1, phi2):
    diff = np.abs(phi1 - phi2)
    if diff > 180.0:
        return 360.0 - diff
    else:
        return diff

def deg_min_to_dec(degrees, minutes):
    return degrees + minutes / 60.0

lambda_val = 632.8e-9

# Dane z tabeli
phi_1L = np.array([deg_min_to_dec(353, 40), deg_min_to_dec(351, 5), deg_min_to_dec(348, 20), deg_min_to_dec(345, 35)])
phi_2L = np.array([deg_min_to_dec(173, 50), deg_min_to_dec(171, 5), deg_min_to_dec(168, 20), deg_min_to_dec(165, 35)])
phi_1P = np.array([np.nan, deg_min_to_dec(359, 10), deg_min_to_dec(1, 50), deg_min_to_dec(4, 39)])
phi_2P = np.array([np.nan, deg_min_to_dec(179, 6), deg_min_to_dec(181, 35), deg_min_to_dec(184, 39)])

n_indices = np.array([1, 2, 3])

# Wyciągamy wycinki dla n=1,2,3
p1L = phi_1L[1:4]
p2L = phi_2L[1:4]
p1P = phi_1P[1:4]
p2P = phi_2P[1:4]

phi_values_deg = []
for i in range(len(n_indices)):
    sep_1 = angle_diff(p1L[i], p1P[i])
    sep_2 = angle_diff(p2L[i], p2P[i])
    # UWAGA: Aby uzyskać wyniki zgodne z poprzednim autorem (~4.5um),
    # traktujemy pełną separację jako kąt do wzoru (lub siatka jest gęstsza).
    # Dzielimy przez 2.0 (średnia z dwóch noniuszy), ale NIE dzielimy na pół dla kąta ugięcia.
    phi_mean = (sep_1 + sep_2) / 2.0 
    phi_values_deg.append(phi_mean)

phi_values_deg = np.array(phi_values_deg)
phi_values_rad = np.deg2rad(phi_values_deg)

# Obliczenie stałej siatki
d_values = (n_indices * lambda_val) / np.sin(phi_values_rad)

# Niepewności
u_phi_deg = (1.0 / 60.0) / np.sqrt(3) # błąd noniusza
u_phi_rad = np.deg2rad(u_phi_deg)

# Propagacja błędu dla d
u_d_values = d_values * (u_phi_rad / np.tan(phi_values_rad)) # ctg(phi)

# Średnia i jej niepewność
d_mean = np.mean(d_values)
u_d_mean_A = np.std(d_values, ddof=1) / np.sqrt(len(d_values))

# Liczba rys na mm
N_rys = 1.0 / d_mean
u_N = N_rys * (u_d_mean_A / d_mean)

print("Kąty (użyte do wzoru) [deg]:", np.round(phi_values_deg, 3))
print("Stałe siatki d [m]:", d_values)
pjint(f"Liczba rys N: {N_rys/1000:.2f} /mm")
print("-" * 30)
print(f"Niepewność u(phi) [deg]: {u_phi_deg:.4f}")
print("Niepewności u(d) [m]:", u_d_values)
print("-" * 30)
print(f"Średnia stała siatki d_sr: {d_mean:.4e} m")
print(f"Niepewność u(d_sr) (Typ A): {u_d_mean:.4e} m")
print("-" * 30)
print(f"Liczba rys na 1 mm: {N_rys/1000:.2f}")
print(f"Niepewność liczby rys [mm^-1]: {u_N/1000:.2f}")
`````
