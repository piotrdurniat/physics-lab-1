## Zadanie 2 - uproszczenie wzorów:

```python
import sympy as sp

# 1. Definiowanie symboli
E, R1, R2, R3, R4 = sp.symbols('E R1 R2 R3 R4')

# 2. Definiowanie "surowych" równań fizycznych (z ułamkami piętrowymi)
# Opór zastępczy gałęzi równoległej (R2 || (R3+R4))
R_parallel = (R2 * (R3 + R4)) / (R2 + R3 + R4)

# Opór zastępczy całego układu
Rz = R1 + R_parallel

# Prąd I1 (Prawo Ohma)
I1_raw = E / Rz

# Prąd I3 (Dzielnik prądu) - I3 płynie przez gałąź (R3+R4)
I3_raw = I1_raw * (R2 / (R2 + R3 + R4))
I4_raw = I3_raw # Są w szeregu

# Prąd I2 (I prawo Kirchhoffa: I2 = I1 - I3)
I2_raw = I1_raw - I3_raw

# 3. Upraszczanie wyrażeń (sprowadzanie do najprostszej postaci)
print("--- UPROSZCZONE WYNIKI ---")

print("\n1. Prąd I1:")
I1_simple = sp.simplify(I1_raw)
sp.pprint(I1_simple)
print("LaTeX:", sp.latex(I1_simple))

print("\n2. Prąd I2:")
I2_simple = sp.simplify(I2_raw)
sp.pprint(I2_simple)
print("LaTeX:", sp.latex(I2_simple))

print("\n3. Prąd I3 (oraz I4):")
I3_simple = sp.simplify(I3_raw)
sp.pprint(I3_simple)
print("LaTeX:", sp.latex(I3_simple))

# 4. Weryfikacja wspólnego mianownika (Wyznacznika)
# Sprawdzamy czy mianownik jest taki sam dla wszystkich
denom_I1 = sp.denom(I1_simple)
denom_I2 = sp.denom(I2_simple)
denom_I3 = sp.denom(I3_simple)

print("\n--- WERYFIKACJA MIANOWNIKA ---")
if denom_I1 == denom_I2 == denom_I3:
    print("Sukces: Wszystkie wzory mają ten sam wspólny mianownik:")
    sp.pprint(denom_I1)
else:
    print("Uwaga: Mianowniki się różnią.")
```
